### Checking the performance of CSV v/s parquet formats

```python
import pandas as pd
from pathlib import Path

DATA_PATH_CSV = Path("../data/processed/equity/master/nse_master_2016_2026.csv")
df = pd.read_csv(DATA_PATH_CSV)
df = df.sort_values(["symbol","trade_date"])
df
```

```python
import pandas as pd
from pathlib import Path

DATA_PATH = Path("../data/processed/equity/parquet/nse_master_2016_2026.parquet")
df = pd.read_parquet(DATA_PATH)
df = df.sort_values(["symbol","trade_date"])
df
```

<b>Note:</b><i>From above example we can see the difference between csv & parquet. That, <b>Parquet is more faster than csv</b></i>

## Daily Return

```python
df["return_1d"] = df.groupby("symbol")["close"].pct_change()
df
```

## Short Momentum

```python
df["return_5d"] = df.groupby("symbol")["close"].pct_change(5)
df
```

## Medium Momentum

```python
df["return_20d"] = df.groupby("symbol")["close"].pct_change(20)
df
```

## Rolling Volatility

```python
df["volatility_20d"] = (
    df.groupby("symbol")["return_1d"]
      .rolling(20)
      .std()
      .reset_index(level=0, drop=True)
)
df
```

## Liquidity Features

### Traded Value

```python
df["traded_value"] = df["close"] * df["tottrdqty"]
df
```

### Rolling liquidity

```python
df["avg_value_20d"] = (
    df.groupby("symbol")["traded_value"]
      .rolling(20)
      .mean()
      .reset_index(level=0, drop=True)
)
df
```

## Applying Liquidity Filters

```python
LIQUIDITY_THRESHOLD = 5e7
```

(₹5 crore average daily traded value)

### Filter dataset: clean tradable universe

```python
df = df[df["avg_value_20d"] > LIQUIDITY_THRESHOLD]
df = df[df["close"] > 20]
```

### Data after Liquidity Filters

## Relative Strength vs Index

Load index features

```python
path = Path("../data/features/index")

files = sorted(path.glob("index_features_*.parquet"))

index_df = pd.concat(
    [pd.read_parquet(f) for f in files],
    ignore_index=True
)

index_df
```

Merging equity & index features

```python
df = df.merge(
    index_df[["trade_date","return_1d"]],
    on="trade_date",
    suffixes=("","_index")
)
```

Merged Data

```python
df
```

### Compute relative strength

```python
df["relative_strength"] = df["return_1d"] - df["return_1d_index"]
df
```

## Cross-Sectional Ranking

daily ranking across stocks

```python
df["rank_momentum"] = (
    df.groupby("trade_date")["return_20d"]
      .rank(ascending=False)
)
```

```python
df
```

return_20d = (Current Price - Price 20 days ago) / Price 20 days ago
</br>
| Symbol   | Price Today | Price 20d Ago | return_20d |
| -------- | ----------- | ------------- | ---------- |
| RELIANCE | 2600        | 2400          | 8.33%      |
| TCS      | 3800        | 3700          | 2.70%      |
| INFY     | 1600        | 1500          | 6.67%      |

</br>

| symbol   | return_20d | rank_momentum |
| -------- | ---------- | ------------- |
| RELIANCE | 8.33       | 1             |
| INFY     | 6.67       | 2             |
| TCS      | 2.70       | 3             |

Normalize ranking

```python
df["rank_pct"] = (
    df.groupby("trade_date")["return_20d"]
      .rank(pct=True)
)
```

```python
df
```

Important detail: .rank(pct=True) ranks in ascending order by default.
</br>
| symbol   | return_20d |
| -------- | ---------- |
| TCS      | 2.70       |
| INFY     | 6.67       |
| RELIANCE | 8.33       |

</br>

| symbol   | return_20d | rank |
| -------- | ---------- | ---- |
| TCS      | 2.70       | 1    |
| INFY     | 6.67       | 2    |
| RELIANCE | 8.33       | 3    |


rank_pct = (rank) / (total stocks)

</br>

| symbol   | rank | rank_pct       |
| -------- | ---- | -------------- |
| TCS      | 1    | 1/3 = **0.33** |
| INFY     | 2    | 2/3 = **0.67** |
| RELIANCE | 3    | 3/3 = **1.00** |


</br>

| symbol   | return_20d | rank_momentum | rank_pct |
| -------- | ---------- | ------------- | -------- |
| RELIANCE | 8.33       | 1             | **1.00** |
| INFY     | 6.67       | 2             | **0.67** |
| TCS      | 2.70       | 3             | **0.33** |

</br>

rank_pct tells how strong a stock’s momentum is relative to all other stocks on that day.
</br>
<ul>
<li>1.0 → best performer</li>

<li>0.5 → average</li>

<li>0.1 → poor performer</li>
</ul>

## Save Features (Partitioned Parquet)

```python
df["year"] = df["trade_date"].dt.year
df
```

```python
import sys
from pathlib import Path

# Add project root directory to sys.path
sys.path.append(str(Path().resolve().parents[0]))

# importing get_logger function from scripts.logger
from scripts.logger import get_logger

logger = get_logger(__name__)

output_path = Path("../data/features/equity")

# Create equity if not exists
if not output_path.exists():
    output_path.mkdir(parents=True, exist_ok=True)
    logger.info(f"Created directory: {output_path}")
else:
    logger.info(f"Directory already exisits:{output_path}")

years = df["year"].unique()

for year in years:
    try:
    
        df_year = df[df["year"] == year]
        
        file_path = output_path / f"equity_features_{year}.parquet"
        
        # Check if parquet already exists
        if file_path.exists():
            logger.info(f"File already exists for year {year}: {file_path}")
        else:
            df_year.to_parquet(file_path, index=False)
            logger.info(f"Parquet file created for year {year}: {file_path}")

    except Exception as e:
        logger.error(f"Error while creating parquet for year {year}: {e}")
```
