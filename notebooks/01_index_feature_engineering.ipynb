## Daily Return

```python
df["return_1d"] = df["closing_index_value"].pct_change()
df
```

pct_change() calculates the percentage change between the current value and the previous value.</br>
Return=(Current Price−Previous Price​)/Previous Price</br>
If the Result is 0.05 then the returns are 0.05*100 = 5% High from last closing

## Weekly Momentum

```python
df["return_5d"] = df["closing_index_value"].pct_change(5)
df
```

## Monthly Momentum

```python
df["return_20d"] = df["closing_index_value"].pct_change(20)
df
```

## Rolling Volatility

```python
df["volatility_20d"] = df["return_1d"].rolling(20).std()
df
```

moving window of 20 rows (20 days)</br>
looks at the previous 20 days of returns_1d</br>
std() calculates the standard deviation of the values inside the window</br>
In simple words, How much the market has been fluctuating over the last 20 days</br>
Higher std → more volatile market</br>
Lower std → stable market</br>
| Volatility (`volatility_20d`) | Approx %    | Market Type | Meaning                          |
| ----------------------------- | ----------- | ----------- | -------------------------------- |
| **< 0.007**                   | < 0.7%      | Low         | Stable / trending market         |
| **0.007 – 0.015**             | 0.7% – 1.5% | Medium      | Normal market movement           |
| **> 0.015**                   | > 1.5%      | High        | Panic / crash / highly uncertain |


## Drawdown

```python
rolling_max = df["closing_index_value"].cummax()
df["drawdown"] = df["closing_index_value"] / rolling_max - 1
df
```

cummax() means cumulative maximum</br>
(highest value the index has reached so far)</br>
Drawdown = (Current Price​)/(Peak Price) - 1</br>
how far the price has fallen from its highest point</br>
If drawdown is -0.028 then -2.8% drawdown (Meaning price is 2.8% below its peak)

## Market Breadth Features (From Equity Dataset)

Market breadth measures how many stocks are moving up or down in the market.
It helps identify market regimes such as broad rallies or weak participation.

We compute:
- Percentage of stocks with positive daily return

## Load the Equity Dataset

```python
# Equity dataset path
equity_path = Path("../data/processed/equity/parquet/nse_master_2016_2026.parquet")

equity = pd.read_parquet(equity_path)

equity.head()
```

## Compute Daily Returns for Each Stock

```python
# sort values by symbol and trade_date
equity = equity.sort_values(["symbol", "trade_date"])

# daily return per stock
equity["return_1d"] = equity.groupby("symbol")["close"].pct_change()

equity.head()
```

We calculate daily returns for each stock using percentage change.
This allows us to determine whether each stock moved up or down on a given day.

## Compute Breadth Signal

```python
breadth = equity.groupby("trade_date")["return_1d"].apply(
    lambda x: (x > 0).sum() / len(x)
)

breadth = breadth.rename("breadth_positive").reset_index()

breadth.head()
```

Breadth Positive = Number of stocks with positive return / Total stocks

Values close to:
1.0 → Broad market rally  
0.5 → Mixed market  
0.0 → Broad market decline

## Merge Breadth with Index Dataset

```python
df = df.merge(breadth, on="trade_date", how="left")

df
```

## Save Features (Partitioned Parquet)

```python
df["year"] = df["trade_date"].dt.year
df
```

```python
import sys

# Add project root directory to sys.path
sys.path.append(str(Path().resolve().parents[0]))

# importing get_logger function from scripts.logger
from scripts.logger import get_logger

logger = get_logger(__name__)

output_path = Path("../data/features/index")

# Create index if not exists
if not output_path.exists():
    output_path.mkdir(parents=True, exist_ok=True)
    logger.info(f"Created directory: {output_path}")
else:
    logger.info(f"Directory already exisits:{output_path}")

years = df["year"].unique()

for year in years:
    try:
    
        df_year = df[df["year"] == year]
        
        file_path = output_path / f"index_features_{year}.parquet"

        # Check if parquet already exists
        if file_path.exists():
            logger.info(f"File already exists for year {year}: {file_path}")
        else:
            df_year.to_parquet(file_path, index=False)
            logger.info(f"Parquet file created for year {year}: {file_path}")
    
    except Exception as e:
        logger.error(f"Error while creating parquet for year {year}: {e}")
    
    
```
